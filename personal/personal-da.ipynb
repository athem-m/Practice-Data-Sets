# Data Visualizations

```python
# Count by Gender and Mean Age
plt.figure(figsize=(20, 8))
plt.suptitle("Overview of Gender Distribution in Dataset", size=20)
plt.subplot(1,2,1)
sns.countplot(x='gender',data=df);
plt.title('Total Number: Male vs Female ')
plt.subplot(1,2,2)
df.groupby('gender').age.mean().plot(kind='bar')
plt.title('Mean Age of Male and Female');
```

```python
# Distribution of age
plt.figure(figsize=(12, 9))
sns.histplot(data=df, x='age', bins=10, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Proportion of customers with and without partners
partner_counts = df['partner'].value_counts()
plt.figure(figsize=(12, 9))
plt.pie(partner_counts, labels=partner_counts.index, autopct='%1.1f%%')
plt.title('Proportion of Customers with and without Partners')
plt.axis('equal')
plt.show()
```

```python
# Proportion of customers with and without dependents
dependents_counts = df['dependents'].value_counts()
plt.figure(figsize=(12, 9))
plt.pie(dependents_counts, labels=dependents_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Customers with and without Dependents')
plt.axis('equal')
plt.show()
```

```python
# Define colors for genders
gender_colors = {'Male': 'red', 'Female': 'blue'}

# Visualization 2: Distribution of age with colors for genders
plt.figure(figsize=(16, 12))
sns.histplot(data=df, x='age', bins=10, kde=True, hue='gender', palette=gender_colors)
plt.title('Distribution of Age by Gender')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend(title='Gender')
plt.show()
```
